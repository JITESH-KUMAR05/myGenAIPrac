## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning 

```python
## missing values
dataset[dataset.isnull().any(axis = 1)]
```

```python
## now we know that we have 1st region till 122 and 2nd from 122 
## so we will create a new column for regions 
## for region 1 we will give 0 and for 2nd we will give 1
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.info()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df
```

```python
df.isnull().sum()
```

```python
## REmoving all the null values
df = df.dropna().reset_index(drop=True)
print(df.head())
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## let's fix these spaces in the column name
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
## Change the required columns as integers datatype
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the other column data type to float
```

```python
objects = [feature for feature in df.columns if df[feature].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## lets now save this cleaned data set into another csv file
df.to_csv('Algerian_forest_fires_clean_dataset.csv',index=False)
```
