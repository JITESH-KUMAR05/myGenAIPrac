```python
import pandas as pd
messages=pd.read_csv('SpamClassifier-master/spam.csv',
                    sep=',',names=["label","message"],header=0, encoding='latin1',usecols=[0,1])
```

```python
messages
```

```python
## Data Cleaning and Preprocessing 
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
ls = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('^[a-zA-Z]',' ',messages['message'][i])
    review =  review.lower()
    review = review.split()
    review = [ls.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## create the bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
### Create Bag of Words model with ngrams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100,binary=True,ngram_range=(3,3))
X = cv.fit_transform(corpus).toarray()
```
